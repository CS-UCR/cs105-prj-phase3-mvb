## Name Changing:

The NBA has gone and will continue to go through team changes. For example, teams will change cities such as the Minneapolis Lakers changed cities to Los Angeles making them the Los Angeles Lakers. 

A completely different but also notable type of NBA team change is when a team switches cities and in turn also switches team name to get a fresh start. An example is the Seattle SuperSonics changing cities to Oklahoma City and changing their name to become the Oklahoma City Thunder. Althogh it seems like the Seattle SuperSonics and the Oklahoma City Thunder are completely different teams, They are actually the same franchise. There are many other franchises throughout the NBA timeline who have the same situation.
